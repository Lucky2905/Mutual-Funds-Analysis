```python
df_1 = df
```

```python
df_1.loc[df_1["Mod. Duration(in yrs)"]=='-',"Mod. Duration(in yrs)"] = np.nan
df_1.loc[df_1["Avg. Maturity(in yrs)"]=='-',"Avg. Maturity(in yrs)"] = np.nan
```

```python
df_1[["Avg. Maturity(in yrs)","Mod. Duration(in yrs)"]]
```

```python
df_1.loc[df_1["RupeeVestRating"]=='Unrated',"RupeeVestRating"] = None
```

```python
df_1.columns = (df_1.columns
                .str.strip()
                .str.replace(r"[()%]","",regex = True)
                .str.replace(r"\s+","_",regex=True)
                .str.replace(r"\.","",regex=True))
```

```python
df_1.head()
```

```python
rename_dict = {
    "AUMin_Rs_cr":"AUM_Cr",
    "ExpenseRatio_":"Expense_Ratio",
    "Return_1_mo": "Return_1M",
    "Return_3_mo": "Return_3M",
    "Return_6_mo": "Return_6M",
    "Return_1_yr": "Return_1Y",
    "Return_2_yrs": "Return_2Y",
    "Return_3_yrs": "Return_3Y",
    "Return_5_yrs": "Return_5Y",
    "Return_10_yrs": "Return_10Y",
    "Avg_Market_Capin_Rs_cr": "Avg_MarketCap_Cr",
    "Avg_Maturityin_yrs": "Avg_Maturity_Yrs",
    "Mod_Durationin_yrs": "Mod_Duration_Yrs",
    "Yield_To_Maturity_": "YTM"
}

df_1.rename(columns=rename_dict, inplace=True)
df_1.head()
```

```python
numcols = ["AUM_Cr", "Expense_Ratio", "NAV", "52_WeekHigh_NAV", "52_WeekLow_NAV",
            "Return_1M", "Return_3M", "Return_6M", "Return_1Y", "Return_2Y",
            "Return_3Y", "Return_5Y", "Return_10Y", 
            "Avg_Maturity_Yrs", "Mod_Duration_Yrs", "YTM"]
for col in numcols:
    df_1[col] = (df_1[col]
               .astype(str)
               .replace("-",np.nan)
               .replace("nan",np.nan)
               .astype(float))
df_1["Inception_Date"] = pd.to_datetime(
    df_1["Inception_Date"].replace(["-", "--", "N/A", "Na", "na", ""], pd.NA),
    errors="coerce"
)
```

```python
df_2 = df_1
```

```python
df_1["Inception_Date"] = pd.to_datetime(df_1["Inception_Date"], format="%d-%b-%Y", errors="coerce")
```

```python
df_2["Fund_Age_Years"] = round((pd.Timestamp.today() - df_1["Inception_Date"]).dt.days / 365,2)
df_2["Fund_Age_Years"]
```

```python

df_1["Return_1Y_vs_2Y"] = round((df_1["Return_1Y"] / df_1["Return_2Y"]),2)
df_1["Return_2Y_vs_3Y"] = round((df_1["Return_2Y"] / df_1["Return_3Y"]),2)
df_2["Return_1Y_vs_3Y"] = round((df_2["Return_1Y"] / df_2["Return_3Y"]),2)
df_2["Return_3Y_vs_5Y"] = round((df_2["Return_3Y"] / df_2["Return_5Y"]),2)
```

```python
df_2[["Return_1Y_vs_3Y","Return_3Y_vs_5Y"]]
```

```python
return_cols = ["Return_1Y","Return_2Y","Return_3Y","Return_5Y","Return_10Y"]
df_2["Avg_Return"] = df_2[return_cols].mean(axis=1).round(2)
```

```python
df_2["Avg_Return"].head(20)
```

```python
# Volatility Proxy (NAV Range %) -> high % range -> more volatile fund.
df_2["NAV_Range_%"] = (
    (df_2["52_WeekHigh_NAV"] - df_2["52_WeekLow_NAV"]) / df_2["52_WeekLow_NAV"] * 100
).round(2)
```

```python
df.head()
```

- When you invest in a mutual fund, the fund company charges an Expense Ratio (management fees, admin charges, distribution costs, etc.).
- Expense Ratio = % of fund assets deducted every year.
- Expense Impact = the effect of this deduction on your returns over time.
- It directly reduces your Net Asset Value (NAV) and hence your gains.

```python
df_2["Expense_vs_Return_1Y"] = round((df_2["Expense_Ratio"]/df_2["Return_1Y"]),2)
```

```python
df_2["AUM_Category"] = pd.cut(
    df_1["AUM_Cr"],
    bins=[0, 500, 5000, 20000, 100000, float("inf")],
    labels=["Small", "Mid", "Large", "Mega", "Giant"]
)
```

```python
df_2["AUM_Category"]
```

```python
# 3-Year Return is Standard in Mutual Fund Analysis
# 1-year return â†’ too short-term, can be highly volatile and influenced by market noise.
# 5- or 10-year return â†’ too long-term, may not reflect the fund managerâ€™s recent performance or expense impact.
# 3-year return â†’ gives a medium-term perspective: balances stability and relevance.
```

```python
df_2["Return_to_Expense"] = (df_2["Return_3Y"] / df_2["Expense_Ratio"]).round(2)
```

```python
df_2[["Fund_Age_Years", "Avg_Return", "NAV_Range_%", "AUM_Category", "Return_to_Expense"]].head()
```

```python

cat_cols = [
    "Funds", "Fund_Manager", "Category", "Fund_Type",
    "Benchmark_Index", "Highest_Sector", "classification",
    "RupeeVestRating", "AUM_Category"
]

for col in cat_cols:
    df_2[col] = (
        df_2[col]
        .astype(str)
        .str.strip()
        .str.replace("-", " ", regex=True)
        .str.replace("_", " ", regex=True)
        .str.replace(":", " - ", regex=True)
        .str.replace(r"\s+", " ", regex=True)   
        .str.title()                            
        .replace(
            ["Nan", "Na", "None", "Not Rated", "Nr", "Unrated", "Null", ""],
            pd.NA
        )
    )

df_2["Category"] = df_2["Category"].fillna("Unknown")
df_2["Fund_Type"] = df_2["Fund_Type"].fillna("Unknown")
df_2["AUM_Category"] = df_2["AUM_Category"].fillna("Unknown")

df_2[cat_cols].head()
```

```python

df_2["Fund_Age_Years"] = (
    (pd.Timestamp.today() - pd.to_datetime(df_2["Inception_Date"], errors="coerce")).dt.days / 365
).round(2)

df_2["NAV_Range_%"] = (
    ((df_2["52_WeekHigh_NAV"] - df_2["52_WeekLow_NAV"]) / df_2["52_WeekLow_NAV"]) * 100
).round(2)

df_2["Return_1Y_vs_2Y"] = (df_2["Return_1Y"] - df_2["Return_2Y"]).round(2)
df_2["Return_2Y_vs_3Y"] = (df_2["Return_2Y"] - df_2["Return_3Y"]).round(2)
df_2["Return_1Y_vs_3Y"] = (df_2["Return_1Y"] - df_2["Return_3Y"]).round(2)
df_2["Return_3Y_vs_5Y"] = (df_2["Return_3Y"] - df_2["Return_5Y"]).round(2)

return_cols = ["Return_1Y", "Return_3Y", "Return_5Y", "Return_10Y"]
df_2["Avg_Return"] = df_2[return_cols].mean(axis=1).round(2)

df_2["Expense_vs_Return_1Y"] = (
    (df_2["Expense_Ratio"] / df_2["Return_1Y"])
).replace([np.inf, -np.inf], np.nan).round(2)

df_2["Return_to_Expense"] = (
    df_2["Return_1Y"] / df_2["Expense_Ratio"]
).replace([np.inf, -np.inf], np.nan).round(2)

df_2["AUM_Category"] = pd.cut(
    df_2["AUM_Cr"],
    bins=[0, 100, 500, 2000, np.inf],
    labels=["Small", "Mid", "Large", "Mega"]
)

df_2[[
    "Funds", "Fund_Age_Years", "NAV_Range_%", "Return_1Y_vs_2Y",
    "Avg_Return", "Expense_vs_Return_1Y", "Return_to_Expense", "AUM_Category"
]].head()
```

```python

df_3 = df_2.copy()

risk_cols = ["Alpha", "Beta", "Sharpe", "Sortino", "Standard_Deviation"]
for col in risk_cols:
    df_3[col] = pd.to_numeric(df_3[col], errors="coerce")

df_3["Risk_Category"] = pd.cut(
    df_3["Standard_Deviation"],
    bins=[-np.inf, 10, 20, np.inf],
    labels=["Low", "Medium", "High"]
)

df_3["Return_to_Risk"] = (df_3["Avg_Return"] / df_3["Standard_Deviation"]).replace([np.inf, -np.inf], np.nan).round(2)

df_3["Positive_Alpha_Beta_Less1"] = ((df_3["Alpha"] > 0) & (df_3["Beta"] < 1))

df_3["Risk_Adjusted_Performance"] = df_3[["Sharpe", "Sortino"]].mean(axis=1).round(2)

df_3["High_Expense_Low_Return"] = ((df_3["Expense_Ratio"] > df_3["Return_1Y"])).astype(int)

df_3[[
    "Funds", "Risk_Category", "Return_to_Risk", 
    "Positive_Alpha_Beta_Less1", "Risk_Adjusted_Performance", 
    "High_Expense_Low_Return"
]].head(10)
```

```python

df_3["High_Expense_Low_Return"] = (df_3["Expense_Ratio"] > df_3["Return_1Y"]).astype(int)

df_3["High_NAV_Volatility"] = (df_3["NAV_Range_%"] > 20).astype(int)

avg_return = df_3["Avg_Return"].mean()
df_3["Top_Performer"] = (df_3["Avg_Return"] > avg_return).astype(int)

df_3["Young_Fund"] = (df_3["Fund_Age_Years"] < 5).astype(int)

df_3["Efficient_Fund"] = (df_3["Return_to_Expense"] > 5).astype(int)

final_cols = [
    "Funds", "Fund_Manager", "Fund_Age_Years", "NAV_Range_%", "Avg_Return",
    "Expense_Ratio", "Return_to_Expense", "High_Expense_Low_Return",
    "High_NAV_Volatility", "Top_Performer", "Young_Fund", "Efficient_Fund"
]

df_3_final = df_3[final_cols]

df_3_final.head()
```

```python
import numpy as np
import pandas as pd

df_3["Inception_Date"] = pd.to_datetime(df_3["Inception_Date"], errors="coerce")

df_3["ShortTerm_Return"] = df_3[["Return_1M", "Return_3M", "Return_6M"]].mean(axis=1).round(2)

df_3["LongTerm_Return"] = df_3[["Return_3Y", "Return_5Y", "Return_10Y"]].mean(axis=1).round(2)

df_3["Risk_Category"] = pd.cut(
    df_3["Standard_Deviation"],
    bins=[-np.inf, 5, 15, np.inf],
    labels=["Low", "Medium", "High"]
)

df_3["Return_to_Risk"] = (df_3["Return_3Y"] / df_3["Standard_Deviation"]).replace([np.inf, -np.inf], np.nan).round(2)

fund_overview = df_3[["Funds", "Fund_Manager", "Category", "AUM_Cr", "Expense_Ratio"]]

performance_metrics = df_3[[
    "Funds", "Return_1M", "Return_3M", "Return_6M", "Return_1Y",
    "Return_2Y", "Return_3Y", "Return_5Y", "Return_10Y", "Alpha", "Beta", "Sharpe"
]]

risk_profile = df_3[[
    "Funds", "Standard_Deviation", "Turnover_Ratio_", "Avg_MarketCap_Cr", "Large_Cap", "Mid_Cap", "Small_Cap"
]]

portfolio_allocation = df_3[[
    "Funds", "Large_Cap", "Mid_Cap", "Small_Cap", "Highest_Sector"
]]

df_3.head()
```

```python
cols_to_drop = [
    "RupeeVestRating", "Exit_load_Remarks", "classification", 
    "No_ofStocks", "Avg_MarketCap_Cr", "Turnover_Ratio_", 
    "Avg_Maturity_Yrs", "Mod_Duration_Yrs", "YTM", 
    "NAV", "52_WeekHigh_NAV", "52_WeekLow_NAV", 
    "Return_1Y_vs_2Y", "Return_2Y_vs_3Y", "Return_1Y_vs_3Y", "Return_3Y_vs_5Y",
    "NAV_Range_%"
]

df_cleaned = df_3.drop(columns=cols_to_drop, errors='ignore')

df_cleaned.to_csv("Mutual_Funds_Visualization_Ready.csv", index=False)
```
